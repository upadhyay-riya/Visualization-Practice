```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

happiness_df = pd.read_csv('https://raw.githubusercontent.com/siddhantbhattarai/Machine_Learning/main/Datasets/World-Happiness-Report.csv')
happiness_df.head()
```

1. Plot the distribution of the Happiness Score:

```python
plt.figure(figsize=(10, 6))
plt.hist(happiness_df['Happiness Score'], bins=20, color='skyblue', edgecolor='black')
plt.title('Happiness Score Distribution')
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.show()
```

2. Create a bar chart showing the top 10 happiest countries:

```python
top_10_happiest = happiness_df.sort_values(by='Happiness Score', ascending=False).head(10)
plt.figure(figsize=(12, 6))
plt.bar(top_10_happiest['Country'], top_10_happiest['Happiness Score'], color='green')
plt.title('Top 10 Happiest Countries')
plt.xlabel('Country')
plt.ylabel('Happiness Score')
plt.xticks(rotation=45)
plt.show()
```

3. Scatter plot the Economy (GDP per capita) against the Happiness Score:

```python
plt.figure(figsize=(10, 6))
plt.scatter(happiness_df['Economy'], happiness_df['Happiness Score'], color='purple')
plt.title('Economy vs. Happiness Score')
plt.xlabel('Economy (GDP per Capita)')
plt.ylabel('Happiness Score')
plt.grid(True)
plt.show()
```

18. Create a stacked bar chart for the top 5 happiest countries showing Economy, Family, and Health scores:

```python
top_5_happiest = happiness_df.sort_values(by='Happiness Score', ascending=False).head(5)
bar_width = 0.4
bar_positions = range(len(top_5_happiest))

plt.figure(figsize=(14, 8))
plt.bar(bar_positions, top_5_happiest['Economy'], width=bar_width, label='Economy')
plt.bar(bar_positions, top_5_happiest['Family'], width=bar_width, bottom=top_5_happiest['Economy'], label='Family')
plt.bar(bar_positions, top_5_happiest['Health'], width=bar_width, 
        bottom=top_5_happiest['Economy'] + top_5_happiest['Family'], label='Health')

plt.xticks(bar_positions, top_5_happiest['Country'], rotation=45)
plt.xlabel('Country')
plt.ylabel('Scores')
plt.title('Economy, Family, and Health Scores in Top 5 Happiest Countries')
plt.legend()
plt.show()
```

21. 3D scatter plot of Economy, Health, and Happiness Score

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(happiness_df['Economy'], happiness_df['Health'], happiness_df['Happiness Score'],
                     c=happiness_df['Happiness Score'], cmap='viridis', alpha=0.6)

ax.set_title('3D Scatter Plot of Economy, Health, and Happiness Score')
ax.set_xlabel('Economy')
ax.set_ylabel('Health')
ax.set_zlabel('Happiness Score')
fig.colorbar(scatter, ax=ax)
plt.show()
```

23. Complex subplots combining different plot types (line plot, scatter plot, histogram, and box plot)

```python
# 23. Complex subplots combining different plot types (line plot, scatter plot, histogram, and box plot)
fig, axs = plt.subplots(2, 2, figsize=(14, 12))

# Subplot 1: Line plot
axs[0, 0].plot(happiness_df['Happiness Rank'], happiness_df['Happiness Score'], color='blue')
axs[0, 0].set_title('Happiness Rank vs. Happiness Score')
axs[0, 0].set_xlabel('Happiness Rank')
axs[0, 0].set_ylabel('Happiness Score')

# Subplot 2: Scatter plot
axs[0, 1].scatter(happiness_df['Economy'], happiness_df['Happiness Score'], color='purple')
axs[0, 1].set_title('Economy vs. Happiness Score')
axs[0, 1].set_xlabel('Economy')
axs[0, 1].set_ylabel('Happiness Score')

# Subplot 3: Histogram
axs[1, 0].hist(happiness_df['Family'], bins=20, color='skyblue', edgecolor='black')
axs[1, 0].set_title('Family Scores Distribution')
axs[1, 0].set_xlabel('Family')
axs[1, 0].set_ylabel('Frequency')

# Subplot 4: Box plot
sns.boxplot(x='Region', y='Happiness Score', data=happiness_df, ax=axs[1, 1])
axs[1, 1].set_title('Happiness Scores by Region')
axs[1, 1].set_xlabel('Region')
axs[1, 1].set_ylabel('Happiness Score')
axs[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

24. Create a diverging bar chart showing the difference between Family and Health scores for the top 5 happiest countries

```python
# 24. Create a diverging bar chart showing the difference between Family and Health scores for the top 5 happiest countries
top_5_happiest = happiness_df.sort_values(by='Happiness Score', ascending=False).head(5)
top_5_happiest['Difference'] = top_5_happiest['Family'] - top_5_happiest['Health']
plt.figure(figsize=(12, 6))
plt.barh(top_5_happiest['Country'], top_5_happiest['Difference'], color=['green' if x > 0 else 'red' for x in top_5_happiest['Difference']])
plt.title('Difference Between Family and Health Scores for Top 5 Happiest Countries')
plt.xlabel('Difference (Family - Health)')
plt.ylabel('Country')
plt.gca().invert_yaxis()
plt.show()
```
